# Análise dos dados da Copa do Mundo
Com base na página sobre a Copa do Mundo na Wikipedia vamos obter os dados de público pagante em cada final

link: https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA

```python
# Importando as bibliotecas
import pandas as pd
import requests
# Importando a biblioteca para gráficos
# import seaborn as sns
# import matplotlib.pyplot as plt
```

```python
# URL
url = 'https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA'

# Requisitando a página e guardando a resposta
resposta = requests.get(url)

# Exibir
#resposta.text

# Obter a resposta em formato HTML
table = pd.read_html(resposta.text)

# Dando uma olhada no resultado
table
```

```python
# Apenas a tabela com os pagantes por final
df = table[11]
df.head()
```

```python
# Converter a coluna público pagante para int
# primeiro transforma em string(str) removendo o ponto - > .str.replace('.', '')
# e depois transforma para int - > .astype(int)
# df['Público pagante'] = df['Público pagante'].str.replace('.', '').astype(int)
```

```python
df['Público pagante'] = df['Público pagante'].astype(int)
```

```python
# Construir um gráfico a partir da evolução do público pagante

# Importando a biblioteca para gráficos
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Serve para exibir o gráfico no notebook
%matplotlib inline
```

```python
# Exibindo o gráfico de linhas com marcação em cada ponto
# data=df -> pega os dados do dataframe
sns.pointplot(x='Ano', y='Público pagante', data=df)

# Ajustando as legendas
plt.xticks(rotation=65)
plt.show()
```
