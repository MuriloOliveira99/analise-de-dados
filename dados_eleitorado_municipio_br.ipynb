# Análise de Dados do Eleitorado Brasileiro

```python
# Importando as bibliotecas
import pandas as pd

# Importando as bibliotecas gráficas
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('eleitorado_municipio_2020.csv', encoding='latin1', delimiter=";")
df.head()
```

```python
# .T -> serve para inverter colunas x linhas para uma melhor visualização
df.head(3).T
```

```python
# Tipos e quantidades dos dados
df.info()
```

```python
# As 5 cidades com maior quantidade de eleitores deficientes
df.nlargest(5, 'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Eleitoras: ', df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores: ', df['QTD_ELEITORES_MASCULINO'].sum())
print('Não Informado: ', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Varíaveis para calcular o percentual
total_eleitores = df['QTD_ELEITORES'].sum()
total_feminino = df['QTD_ELEITORES_FEMININO'].sum()
total_masculino = df['QTD_ELEITORES_MASCULINO'].sum()
total_nao_informado = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
# Mostrar os valores percentualmente
print('Eleitoras: ', ((total_feminino / total_eleitores) * 100).round(2))
print('Eleitores: ', ((total_masculino / total_eleitores) * 100).round(2))
print('Não Informado: ', ((total_nao_informado / total_eleitores) * 100).round(2))
```

```python
# Quantos municipios tem mais homens que mulheres?
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Coluna nova para indicar a relação fem/masc
df['RELACAO_FM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# Quais são os municipios com maior relação F/M
df.nlargest(5, 'RELACAO_FM').T
```

```python
# Quais os municipios com menor relação f/m
df.nsmallest(5, 'RELACAO_FM').T
```

```python
# Quais os municipios com menor relação f/m
df.nsmallest(5, 'RELACAO_FM').T
```

```python
# Gráfico de distribuição da relação f/m
# bins=valor = quantidade de barras
df['RELACAO_FM'].plot.hist(bins=20)
```

```python
# Gráfico de distribuição da relação f/m
# usando seaborn
#histplot - Nome da coluna, qtd.Linhas, cor e kde(grafico de aproximacao)
sns.histplot(df['RELACAO_FM'], bins=100, color='red', kde=True)

# Embelezando o gráfico
plt.title('Relação Eleitoras/Eleitores', fontsize=18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.axvline(1.0, color='black', linestyle='--')
```

```python
# Verificar a distribuição f/m mas mostrando os pontos (municipios)
# sns.swarmplot(data=df, x='NM_REGIAO', y='RELACAO_FM')
sns.stripplot(data=df, x='NM_REGIAO', y='RELACAO_FM')
plt.axhline(1.0, color='black', linestyle='--')
```
