```python
# Importando as bibliotecas
import pandas as pd
```

```python
df = pd.read_csv('reembolso_senadores_2018.csv', delimiter=';', encoding='latin1', skiprows=1, decimal=',')
df.head()
```

```python
# Tamanho e tipos dos dados
df.info()
# df.shape
```

```python
df.dtypes
```

```python
# Análise quantitativa dos dados
df.describe()
```

```python
# Qual o total de reembolso?
df['VALOR_REEMBOLSADO'].sum()
```

```python
# Quantos reembolsos foram solicitados por cada senador?
df['SENADOR'].value_counts()
```

```python
# Quanto cada senador solicitou de reembolso em 2018?
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
# Quais são os 5 maiores valores reembolsado?
# T inverte coluna e linha
df.nlargest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Quais são os 5 menores valores reembolsados?
df.nsmallest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Quantidade de reembolso por tipos de despesa
df['TIPO_DESPESA'].value_counts()
```

```python
# Quanto foi reembolsado por cada tipo de despesa
df.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```
